```python
##normalize 'Memory' column
df['Memory'] = df['Memory'].astype(str).replace('\.0', '')
df['Memory'] = df['Memory'].str.replace('GB', '')
df['Memory'] = df['Memory'].str.replace('TB', '000')

```

```python
# #Split into two parts 
memory_split = df['Memory'].str.split('+', expand=True)
df['first'] = memory_split[0].str.strip()
df['second'] = memory_split[1].fillna('0').str.strip()

```

```python
# implementing function
def get_storage_info(mem_str, storage_type):
    return int(''.join(filter(str.isdigit, mem_str))) if storage_type in mem_str else 0


```

```python
##Applying function to both first and second parts

df['HDD'] = df.apply(lambda row: get_storage_info(row['first'], 'HDD') + get_storage_info(row['second'], 'HDD'), axis=1)
df['SSD'] = df.apply(lambda row: get_storage_info(row['first'], 'SSD') + get_storage_info(row['second'], 'SSD'), axis=1)
df['Hybrid'] = df.apply(lambda row: get_storage_info(row['first'], 'Hybrid') + get_storage_info(row['second'], 'Hybrid'), axis=1)
df['Flash_Storage'] = df.apply(lambda row: get_storage_info(row['first'], 'Flash Storage') + get_storage_info(row['second'], 'Flash Storage'), axis=1)

```

MODEL SELECTION

```python
from sklearn.model_selection import train_test_split

x=df.drop(['Price_euros'],axis=1)
y=df['Price_euros']
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
categorical_columns=x.select_dtypes(include='object').columns
numerical_columns=x.select_dtypes(exclude='object').columns
```

```python
numeric_transformer=StandardScaler()
oh_transformer=OneHotEncoder(handle_unknown='ignore',drop='first')
```

```python
preprocessor=ColumnTransformer(
    [
         ("OneHotEncoder",oh_transformer,categorical_columns),
         ("StandardScaler",numeric_transformer,numerical_columns)
    ],remainder='passthrough'
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
```

```python
training=Pipeline([
    ('preprocessing', preprocessor),
    ('regressor',LinearRegression())
])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
training.fit(x_train,y_train)
```

```python
y_pred=training.predict(x_test)
```

```python
x_train
```

```python
y_pred
```

```python
## Evaluation

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2=r2_score(y_test,y_pred)
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
print(f"R2 Score:{r2:.2f}")
print(f"RMSE:{rmse:.2f}")
```
